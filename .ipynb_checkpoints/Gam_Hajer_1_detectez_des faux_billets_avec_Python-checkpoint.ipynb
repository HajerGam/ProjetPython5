## 2 réalisation d'une regression lineaire multiples pour imputer les valeurs manquantes 

## 2.1 régression linéaire multiple avec statsmodels

```python
# Encoder la variable is_genuine
fichier_billet['is_genuine'].replace([False,True], [0,1], inplace=True)
```

```python
#Création d'un dataframe sans les valeures manquantes 
data=fichier_billet.loc[fichier_billet['margin_low'].notnull(),:]
data.head()
```

```python
#Création d'un data frame avec les valeures manquantes 
data_manq=fichier_billet.loc[fichier_billet['margin_low'].isnull(),:]
data_manq.head()
```

```python
#Effectuer une regression linéaire multiple avec statsmodels
reg_multi = smf.ols('margin_low~diagonal+height_left+height_right+margin_up+length', data=data).fit()
print(reg_multi.summary())
```

Evaluation du modele
==>significativité global du modéle avec test de Fisher:
Prob(F-Statistic):elle est très faible et inférieure au seuil alpha 5 % , on rejet l'hypothèse nulle, notre modèle est significatif.
==> test de significativité sur chacune des variables avec test de student:
les cinqs variables diagonal,height_left,height_right,margin_up, margin_up et length sont significatvent avec un p_value =0 inferieur seuil alpha 5%
==> 𝑅2 c’est le pourcentage de variation de la variable dépendante qui est expliqué par le modèle, il vaut environ 0.47, tout comme le  𝑅2ajusté.

 

```python
data_final.info()
```

# Vérifier la colinéarité des variables

```python
#testez la colinearité
variables = reg_multi.model.exog
VIF=[variance_inflation_factor(variables, i) for i in np.arange(1,variables.shape[1])]
VIF
```

==>VIF (acteur d'influence de la variance)inferieure a 10 pour tout les variables , les variables ne sont colinéaire 

# Testez l’homoscédasticité

```python
#Testez l’homoscédasticité
_, pval, __, f_pval = statsmodels.stats.diagnostic.het_breuschpagan(reg_multi.resid, variables)
print('p value test Breusch Pagan:', pval)
```

==>l'homoscédasticité des résidus, c'est-à-dire la constance de leur variance,
La p-valeur ici est inférieure à  5%, on  rejette l'hypothèse  𝐻0 selon laquelle les variances sont constantes (l'hypothèse d’homoscédasticité).

# Testez la normalité des résidus

```python
#Test shapiro pour vérifier la normalité des residus
shapiro(reg_multi.resid)
```

==> p value inferieure a 5% L'hypothèse nulle est rejetée, les résidus ne suivent pas une loi normale.

```python
#Graphique montre la distribution des residus 
sns.distplot(reg_multi.resid)
```

==>la distribution des résidus est bien symétrique dont la moyenne est proche de 0
==>le fait qu'ils ne soient pas très différents d'une distribution symétrique, et le fait que l'échantillon soit de taille suffisante (supérieure à 30) permettent de dire que les résultats obtenus par le modèle linéaire gaussien ne sont pas absurdes, même si le résidu n'est pas considéré comme étant gaussien.

## 4 création d'un model de prédiction avec régression logistique 

```python
#Audit data_final1
data_final1.describe()
```

```python
#data_y
data_y=data_final1['is_genuine']
```

```python
#data_x
data_x=data_final1.drop('is_genuine',axis=1)
```

```python
#scaler
scaler=StandardScaler()
```

```python
#Normalisation des données
std_scale=scaler.fit(data_x)
data_x_scaled=std_scale.transform(data_x)
data_x_scaled
```

```python
liste_var=['diagonal','height_left','height_right','margin_low','margin_up','length']
```

```python
data_x_scaled=pd.DataFrame(data_x_scaled,columns=liste_var)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data_x_scaled, data_y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

# initialisation et entraînement en une ligne
clf = LogisticRegression(random_state=42).fit(X_train, y_train)
```

## 4.1 tracer l'histogramme des probabilités des prédictions

```python
#Un bon moyen d'analyser les performances d'un modèle de classification est de tracer l'histogramme des probabilités des prédictions.
y_hat_proba = clf.predict_proba(X_test)[:,1]
import seaborn as sns
sns.histplot(y_hat_proba)
```

==>Le modèle est assez confiant de ses prédiction, la plupart des prédictions ont une probabilité proche de 0 ou de 1

## 4.2 Accuracy

l'accuracy mesure la Performance globale du modèle, représente le taux de prédictions qui correspondent exactement aux étiquettes de classes réelles=TP+TN/(FP+FN+TP+TN)

```python
# predictions 
y_prediction = clf.predict(X_test)

from sklearn.metrics import accuracy_score
print("accuracy",accuracy_score(y_test, y_prediction))
```

## 4.3 Matrice de confusion

La matrice de confusion est un résumé des résultats de prédiction pour un problème particulier de classification. 
Elle compare les données réelles pour une variable cible à celles prédites par un modèle

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_prediction)
```

```python
# afficher les : vrais négatifs, vrais positifs, fauxnégatifs, faux positifs
tn, fp, fn, tp = confusion_matrix(y_test, y_prediction).ravel()
tn, fp, fn, tp
print('Vrais négatifs :',tn,"\n"'Vrais positifs :',tp,"\n"'Faux négatifs :',fn,"\n"'Faux positifs :',fp)
```

```python
from sklearn.metrics import precision_score, recall_score, roc_auc_score
print("Precision:",precision_score(y_test, y_prediction))
print("Recall:",recall_score(y_test, y_prediction))
print("ROC-AUC", roc_auc_score(y_test, clf.predict_proba(X_test)[:, 1]))
```

Precision: la capacité d’un modèle à éviter d’étiqueter des échantillons négatifs comme positifs=TP/(TP+FP)

Recall ou Sensibilité: Couverture des observations vraiment positives=TP/TP+FN

Spécificité: Couverture des observations vraiment négatives=TN/TN+FP

## 4.4 courbe ROC

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, clf.predict_proba(X_test)[:,1])

import matplotlib.pyplot as plt
plt.plot(fpr, tpr)
plt.grid()
plt.title("ROC curve")
```

==>La courbe ROC (Receiver Operating Characteristic) représente la sensibilité (TP/TP+FN) en fonction de la spécificité (TN/FP+TN)
L'aire sous la courbe ROC, l'AUC (Area Under the ROC), est une mesure de la qualité de la classification qui varie entre :
 AUC=1/2: dans le pire des cas ;
 AUC=1: dans le meilleur des cas 
 ==>selon notre graphique le model de prédiction effectue une trés bonne classification.
 

## 5 création d'un modéle de prédiction avec l'algorithme du Kmeans

```python
#Affichage data_final1
data_final1.head()
```

```python
dfkmeans=data_final1.loc[:,['diagonal','height_left','height_right','margin_low','margin_up','length']]
dfkmeans.head()
```

```python
z = dfkmeans.values
features = dfkmeans.columns
z
```

```python
kmeans = KMeans(n_clusters=10)
```

```python
#scaler
scaler = StandardScaler()
```

```python
#centrage et à la réduction des données
scaler.fit(z)
```

```python
z_scaled = scaler.transform(z)
```

```python
# Une liste vide pour enregistrer les inerties :  
intertia_list = [ ]

# Notre liste de nombres de clusters : 
k_list = range(1, 10)

# Pour chaque nombre de clusters : 
for k in k_list : 
    
    # On instancie un k-means pour k clusters
    kmeans = KMeans(n_clusters=k)
    
    # On entraine
    kmeans.fit(z_scaled)
    
    # On enregistre l'inertie obtenue : 
    intertia_list.append(kmeans.inertia_)
```

```python
#Graphique affichant l'inertie intraclasse avec la methode du coude 
fig, ax = plt.subplots(1,1,figsize=(12,6))

ax.set_ylabel("intertia")
ax.set_xlabel("n_cluster")

ax = plt.plot(k_list, intertia_list)
```

```python
kmeans = KMeans(n_clusters=2)
model=kmeans.fit(z_scaled)
```
